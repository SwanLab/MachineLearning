```python
import mnist_loader
import network
import math
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
epochs = 2
learning_rate = 3.0
sampling = 25
total_time_vector = []
total_time_per_step = []
total_accuracy = []
total_batch_accuracy = []
mini_batch_vector = [10**i for i in range(3,4)]

for mini_batch_size in mini_batch_vector:
    training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
    net = network.Network([784, 30, 10])
    net.SGD(training_data, epochs, mini_batch_size, learning_rate,sampling_test=sampling,test_data=test_data)
    batch_accuracy = net.get_accuracy()
    time_vector    = net.get_time_vector()
    time_per_step  = net.get_time_per_step()
    accuracy  = net.get_accuracy_per_step()
    total_time_per_step.append(time_per_step)
    total_accuracy.append(accuracy)
    total_time_vector.append(time_vector)
    total_batch_accuracy.append(batch_accuracy)
    print(net.get_total_time())
```

```python
print(total_time_per_step)
total_accuracy = [[0.118,0.2237,0.2815,0.3432]]
for i in range(1,len(mini_batch_vector))
    sampling

print(total_accuracy)
```

```python

plt.figure()
for time_vector, batch_vector, mini_batch_size in zip(total_time_vector, total_batch_accuracy, mini_batch_vector):
    plt.plot(time_vector,[100*batch for batch in batch_vector], label=str(mini_batch_size))
plt.xlabel("Time elapsed [s]")
plt.ylabel("Accuracy [%]")
plt.legend()
plt.savefig("figures/batch_vs_time_nepoch_"+str(epochs)+".png", format="png")
plt.show()
```

```python
plt.figure()
for time_vector, batch_vector, mini_batch_size in zip(total_time_vector, total_batch_accuracy, mini_batch_vector):
    plt.plot(time_vector, label=str(mini_batch_size))
plt.xlabel("Epoch")
plt.ylabel("Time elapsed [s]")
plt.savefig("figures/time_vs_epoch_nepoch_"+str(epochs)+".png", format="png")
plt.legend()
plt.show()
```
